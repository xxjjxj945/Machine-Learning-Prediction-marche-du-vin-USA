## Machine Learning (partie 1 : numérique)
Choisis la métrique la plus adaptée, puis entraine différents modèles/paramètres afin de prédire le prix d'une bouteille en fonction de la note ("points") et de l'année. Evalue les scores et conserve uniquement les meilleurs paramètres. Applique le modèle aux 14 vins du Domaine des Croix pour proposer un prix à chaque bouteille.
```
df_pn_bur=df_pn_bur[np.abs(df_pn_bur['price']-df_pn_bur['price'].mean())<=(3*df_pn_bur['price'].std())]
```

```
X=df_pn_bur[['points','vintage']]
y=df_pn_bur['price']
```

```
X
```

```
X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=42)
```

```
# Try with 3 models: LinearRegression; DecisionTreeRegressor; RandomForestRegressor; KNNRegressor
modelLR = LinearRegression().fit(X_train,y_train)
modelDTR = DecisionTreeRegressor(max_depth = 4).fit(X_train, y_train)  
modelRFR = RandomForestRegressor(max_depth = 4).fit(X_train,y_train)
modelKNNR = KNeighborsRegressor().fit(X_train, y_train)


print("\nLR score train = ",modelLR.score(X_train,y_train))
print("LR score test = ",modelLR.score(X_test,y_test))
print("\nDTR score train = ",modelDTR.score(X_train,y_train))
print("DTR score test = ",modelDTR.score(X_test,y_test))
print("\nRFR score train = ",modelRFR.score(X_train,y_train))
print("RFR score test = ",modelRFR.score(X_test,y_test))
print("\nKNNR score train = ",modelKNNR.score(X_train,y_train))
print("KNNR score test = ",modelKNNR.score(X_test,y_test))
```
